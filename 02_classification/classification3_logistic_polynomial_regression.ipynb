```python
import import_ipynb

from classification2_logistic_regression import Logistic_Regression, standardize
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 读入训练数据
train = np.loadtxt('data/data3.csv', delimiter=',', skiprows=1)
train_x = train[:, 0:2]
train_y = train[:, 2]
```

## 1 Data Cleaning

```python

# 标准化

train_z = standardize(train_x)

# 增加 x0 和 x3


def to_matrix(x):
    x0 = np.ones([x.shape[0], 1])
    # 添加x^2
    x3 = x[:, 0, np.newaxis] ** 2
    return np.hstack([x0, x, x3])


train_X = to_matrix(train_z)
```

## 2 Modeling

```python
model = Logistic_Regression(n=4)
theta = model.fit(train_X, train_y, max_epoch=5000, ETA=1e-3)
```

```python
x1 = np.linspace(-2, 2, 100)
x2 = -(theta[0] + theta[1] * x1 + theta[3] * x1 ** 2) / theta[2]
plt.plot(train_z[train_y == 1, 0], train_z[train_y == 1, 1], 'o')
plt.plot(train_z[train_y == 0, 0], train_z[train_y == 0, 1], 'x')
plt.plot(x1, x2, linestyle='dashed')
plt.show()
```
