```python
# 代码清单 4-2-(2)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x0, x1):
    """
    定义 f
    """
    return (2 * x0**2 + x1**2) * np.exp(-(2 * x0**2 + x1**2))

def df_dx0(x0, x1):
    """ 
    f 对 x0 的偏导数
    """
    return (-8 * x0**3 - 4*x0*x1**2+4*x0) * np.exp(-(2 * x0**2 + x1**2))

def df_dx1(x0, x1):
    """
    f 对 x1 的偏导数
    """
    return (-4 * x1**3 - 4*x1*x0**2+2*x1) * np.exp(-(2 * x0**2 + x1**2))

xn=30
x_range=2
x0 = np.linspace(-x_range, x_range, xn)   
x1 = np.linspace(-x_range, x_range, xn)  

xx0, xx1 = np.meshgrid(x0, x1)

# 求函数、偏导数
ff = f(xx0, xx1) 
dff_dx0 = df_dx0(xx0, xx1) 
dff_dx1 = df_dx1(xx0,xx1)





# 绘制梯度变化
"""
箭头指向斜率最大的地方
"""
fig = plt.figure(figsize=(9, 8))
fig.set_facecolor('white')
plt.quiver(xx0, xx1, dff_dx0, dff_dx1)          
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + 0.5)
plt.ylim(-x_range - 0.5, x_range + 0.5)
plt.savefig('gradient.png',dpi=200)
plt.show()
```

```python
# 指数函数
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-4, 4, 100)
y = 2**x
y2 = 3**x
y3 = 0.5**x

plt.figure(figsize=(5, 5))
plt.plot(x, y, 'black', linewidth=3, label='$y=2^x$')
plt.plot(x, y2, 'cornflowerblue', linewidth=3, label='$y=3^x$')
plt.plot(x, y3, 'gray', linewidth=3, label='$y=0.5^x$')
plt.ylim(-2, 6)
plt.xlim(-4, 4)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
# 对数函数
x = np.linspace(-8, 8, 100)
y = 2**x

x2 = np.linspace(0.001, 8, 100) # np.log(0)会导致出错，所以不能包含 0
y2 = np.log(x2) / np.log(2)# 相当于log2(x)

plt.figure(figsize=(5, 5))
plt.plot(x, y, 'black', linewidth=3)
plt.plot(x2, y2, 'cornflowerblue', linewidth=3)
plt.plot(x, x, 'black', linestyle='--', linewidth=1)
plt.ylim(-8, 8)
plt.xlim(-8, 8)
plt.grid(True)
plt.show()
```

```python
# 对数函数只改变大小，不改变最值
x = np.linspace(-4, 4, 100)
y = (x - 1)**2 + 2
logy = np.log(y)

plt.figure(figsize=(4, 4))
plt.plot(x, y, 'black', linewidth=3)
plt.plot(x, logy, 'cornflowerblue', linewidth=3)
plt.yticks(range(-4,9,1))
plt.xticks(range(-4,5,1))
plt.ylim(-4, 8)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()
```

```python
# 求指数函数的导数 y=2^x ,dy=log(2)*2^x
x = np.linspace(-4, 4, 100)
a = 2
y = a**x
dy = np.log(a) * y

plt.figure(figsize=(4, 4))
plt.plot(x, y, 'gray', linestyle='--', linewidth=3)
plt.plot(x, dy, color='black', linewidth=3)
plt.ylim(-1, 8)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()
```

```python
# softmax
def softmax(x0, x1, x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u

# test
y = softmax(2, 1, -1)
print(np.round(y,2))  # (A) 显示小数点后两位的概率
print(np.sum(y))      # (B) 显示和
```

```python
# 标准高斯函数
def gauss(mu, sigma, a):
    """
    mu: 均值
    sigma: 标准差
    a: 高度，当求概率分布时，a=1/sqrt(2 * PI * sigma**2)
    """
    return a * np.exp(-(x - mu)**2 / (2 * sigma**2))


x = np.linspace(-4, 4, 100)
plt.figure(figsize=(4, 4))
plt.plot(x, gauss(0, 1, 1), 'blue', linewidth=3)
plt.plot(x, gauss(2, 2, 0.5), 'red', linewidth=3)
plt.ylim(-.5, 1.5)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()
```

```python
# 二维高斯函数
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

# 高斯函数 -----------------------------
def gauss(x, mu, sigma):
    """
    sigma：协方差
    """
    N, D = x.shape
    c1 = 1 / (2 * np.pi)**(D / 2)
    c2 = 1 / (np.linalg.det(sigma)**(1 / 2)) # sigma的行列式
    a = c1*c2
    inv_sigma = np.linalg.inv(sigma)
    c3 = x - mu
    c4 = np.dot(c3, inv_sigma)
    c5 = np.zeros(N)
    for d in range(D):
        c5 = c5 + c4[:, d] * c3[:, d]
    p = a * np.exp(-c5 / 2)
    return p
```

```python
# 代码清单 4-5-(2)
x = np.array([[1, 2], [2, 1], [3, 4]])
mu = np.array([1, 2])
sigma = np.array([[1, 0], [0, 1]])
print(gauss(x, mu, sigma))
```

```python
# 代码清单 4-5-(3)
X_range0=[-3, 3]
X_range1=[-3, 3]

# 显示等高线 --------------------------------
def show_contour_gauss(mu, sig):
    xn = 40  # 等高线的分辨率
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[xx0.reshape(xn * xn, 1), xx1.reshape(xn * xn, 1)]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    cont = plt.contour(xx0, xx1, f, 15, colors='k')
    plt.grid(True)
    
# 三维图形 ----------------------------------
def show3d_gauss(ax, mu, sig):
    xn = 40  # 等高线的分辨率
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[xx0.reshape(xn * xn, 1), xx1.reshape(xn * xn, 1)]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    f = f.T
    ax.plot_surface(xx0, xx1, f, 
                    rstride=2, cstride=2, alpha=0.3, 
                    color='blue', edgecolor='black')
    
# 主处理 -----------------------------------
mu = np.array([1, 0.5])            # (A)
sigma = np.array([[2, 1], [1, 1]]) # (B)
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1, 2, 1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
Ax = Fig.add_subplot(1, 2, 2, projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
